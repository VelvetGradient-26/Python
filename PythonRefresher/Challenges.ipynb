### Challenge 1: Armstrong Number Checker
Armstrong number: sum of cubes of digits equals the number  
Example: 153 = 1³ + 5³ + 3³ = 1 + 125 + 27 = 153

```python
# TODO: Check if number is Armstrong number
num = int(input("Enter a number: "))
# Your code here

digits = len(str(num))
temp = num
armstrong_sum = 0

while temp > 0:
    digit = temp % 10
    armstrong_sum += digit ** digits
    temp //= 10

if armstrong_sum == num:
    print(f"{num} is an Armstrong number")
else:
    print(f"{num} is not an Armstrong number")
```

### Challenge 3: Perfect Number Finder
Perfect number: sum of proper divisors equals the number  
Example: 6 = 1 + 2 + 3 (divisors of 6)



```python
# TODO: Check if number is perfect
num = int(input("Enter a number: "))
# Your code here
divisor_sum = 0

for i in range(1, num):
    if num % i == 0:
        divisor_sum += i

if divisor_sum == num:
    print(f"{num} is a perfect number")
else:
    print(f"{num} is not a perfect number")
```

### Challenge 4: Star Pattern - Hourglass

```python
# TODO: Create hourglass pattern
# *********
#  *******
#   *****
#    ***
#     *
#    ***
#   *****
#  *******
# *********
n = 5

# Top half
for i in range(n, 0, -1):
    spaces = n - i
    stars = 2 * i - 1
    print(" " * spaces + "*" * stars)

# Bottom half
for i in range(2, n + 1):
    spaces = n - i
    stars = 2 * i - 1
    print(" " * spaces + "*" * stars)
```

### Challenge 1: Remove Duplicates
Given a list with duplicates, create a new list with only unique elements (order preserved).

```python
nums = [1, 2, 3, 2, 4, 1, 5]

seen = set()
unique = []

for num in nums:
    if num not in seen:
        seen.add(num)
        unique.append(num)

print(unique)
```

### Challenge 2: Second Largest Element
Find the second largest unique element in a list of numbers.

```python
nums = [5, 1, 9, 9, 7, 3]

largest = float('-inf')
second = float('-inf')

for num in nums: 
    if num > largest: 
        second = largest
        largest = num
    if num > second and num < largest: 
        second = num

print(second)
```

### Challenge 3: Pair Sum
Given a list of numbers and a target sum, print all pairs that add up to the target.

```python
nums = [1, 2, 3, 4, 5]
target = 6

seen = set()

for num in nums:
    complement = target - num

    if complement in seen:
        print(complement, num)

    seen.add(num)
```

### Challenge 4: Tuple-based Coordinate Distance
Given two points `(x1, y1)` and `(x2, y2)` as tuples, compute the Euclidean distance.

```python
import math

p1 = (0, 0)
p2 = (3, 4)

distance = math.sqrt(
    (p2[0] - p1[0]) ** 2 +
    (p2[1] - p1[1]) ** 2
)

print(distance)
```

## Dictionaries and Sets

### Build a word-frequency analyzer for user-input text


```python
text = input("Enter text: ")

words = text.lower().split()

freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

for word, count in freq.items():
    print(word, ":", count)
```

### Track inventory (item -> quantity, price) with dictionaries


```python
inventory = {}

while True:
    print("\n1. Add Item")
    print("2. View Inventory")
    print("3. Total Inventory Value")
    print("4. Exit")

    choice = input("Choose an option: ")

    match choice:
        case "1":
            item = input("Item name: ")
            quantity = int(input("Quantity: "))
            price = float(input("Price: "))

            inventory[item] = {
                "quantity": quantity,
                "price": price
            }

        case "2":
            print("\nInventory:")

            if not inventory:
                print("Inventory is empty.")
            else:
                for item, details in inventory.items():
                    print(
                        f"{item} | Qty: {details['quantity']} | Price: {details['price']}"
                    )

        case "3":
            total = 0

            for details in inventory.values():
                total += details["quantity"] * details["price"]

            print(f"\nTotal Inventory Value: {total}")

        case "4":
            print("Exiting...")
            break

        case _:
            print("Invalid choice.")
```

### Write a function `is_prime(n)` that returns True/False.


```python
def is_prime(n):
  flag = True

  if n<2:
    flag = False

  if n==2:
    flag = True

  for m in range(2, int(n**(1/2))+1):
    if n % m == 0:
      flag = False
      break

  return flag

var = is_prime(4)
if var:
  print("Number is PRIME")
else:
  print("Number is NOT PRIME")
```

### Write a function `stats(lst)` that returns (min, max, average).

```python
def stats(lst):
    minimum = min(lst)
    maximum = max(lst)
    average = sum(lst) / len(lst)

    return minimum, maximum, average


numbers = [10, 20, 30, 40, 50]

mn, mx, avg = stats(numbers)

print("Min:", mn)
print("Max:", mx)
print("Average:", avg)
```
