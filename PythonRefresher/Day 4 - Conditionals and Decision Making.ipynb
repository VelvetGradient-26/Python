## 1. The if Statement

The `if` statement executes code only when a condition is True.

### Syntax:
```python
if condition:
    # code to execute if condition is True
    # indentation is required!
```

### Key Points:
- Condition must evaluate to True or False
- Colon `:` at the end is required
- Indentation (4 spaces) defines the code block

````python
# Simple if statement
age = 19

if age >= 18:
    print("You are an adult")
    print("You can vote")

print("This always prints")  # Outside if block
````

````python
# Multiple conditions
temperature = 45

if temperature <= 30:
    print("It's hot outside!")
    print("Stay hydrated")

else:
    print("Extremely hot! Stay indoors")
````

## 2. The if-else Statement

Execute one block if condition is True, another if False.

### Syntax:
```python
if condition:
    # code if True
else:
    # code if False
```

````python
# if-else example
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult")
    print("You can vote")
else:
    print("You are a minor")
    print("You cannot vote yet")

print("\nThank you!")
````

````python
# Even or odd checker
number = int(input("Enter a number: "))

if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
````

## 3. The if-elif-else Statement

Check multiple conditions in sequence.

### Syntax:
```python
if condition1:
    # code if condition1 is True
elif condition2:
    # code if condition2 is True
elif condition3:
    # code if condition3 is True
else:
    # code if all conditions are False
```

````python
# Grade assignment
marks = float(input("Enter marks (0-100): "))

if marks >= 90:
    grade = 'A'
    print("Excellent!")
elif marks >= 80:
    grade = 'B'
    print("Very Good!")
elif marks >= 70:
    grade = 'C'
    print("Good")
elif marks >= 60:
    grade = 'D'
    print("Satisfactory")
elif marks >= 40:
    grade = 'E'
    print("Pass")
else:
    grade = 'F'
    print("Fail")

print(f"\nYour grade: {grade}")
````

````python
# Temperature advisory
temp = float(input("Enter temperature (°C): "))

if temp > 40:
    print("Extreme heat! Stay indoors.")
elif temp > 30:
    print("Hot weather. Stay hydrated.")
elif temp > 20:
    print("Pleasant weather.")
elif temp > 10:
    print("Cool weather. Carry a jacket.")
else:
    print("Cold weather. Wear warm clothes.")
````

## 4. Nested if Statements

if statements inside other if statements.

````python
# Nested if example: Loan eligibility
age = int(input("Enter age: "))
income = float(input("Enter monthly income: "))
credit_score = int(input("Enter credit score: "))

print("\n--- Loan Eligibility Check ---")

if age >= 21:
    print("✓ Age requirement met")

    if income >= 30000:
        print("✓ Income requirement met")

        if credit_score >= 700:
            print("✓ Credit score requirement met")
            print("\n🎉 Loan APPROVED!")
        else:
            print("✗ Credit score too low")
            print("\n❌ Loan REJECTED")
    else:
        print("✗ Income insufficient")
        print("\n❌ Loan REJECTED")
else:
    print("✗ Age requirement not met")
    print("\n❌ Loan REJECTED")
````

````python
# Nested if: Number classification
num = int(input("Enter a number: "))

if num > 0:
    print("Positive number")

    if num % 2 == 0:
        print("Even positive number")
    else:
        print("Odd positive number")

elif num < 0:
    print("Negative number")

    if num % 2 == 0:
        print("Even negative number")
    else:
        print("Odd negative number")
else:
    print("Zero (neither positive nor negative)")
````

## 5. Ternary Conditional Operator

Shorthand for simple if-else statements.

### Syntax:
```python
value = true_value if condition else false_value
```

````python
# Traditional if-else
age = 20
if age >= 18:
    status = "Adult"
else:
    status = "Minor"
print(f"Traditional: {status}")

# Ternary operator (same result)
status = "Adult" if age >= 18 else "Minor"
print(f"Ternary: {status}")
````

````python
# More examples
number = 15
parity = "Even" if number % 2 == 0 else "Odd"
print(f"{number} is {parity}")

# With calculations
marks = 75
result = "Pass" if marks >= 40 else "Fail"
print(f"Result: {result}")

# Nested ternary (not recommended for readability)
marks = 85
grade = "A" if marks >= 90 else ("B" if marks >= 80 else "C")
print(f"Grade: {grade}")
````

## 6. Common Pitfalls and Errors

### Error 1: Indentation Errors

````python
# ❌ WRONG: Inconsistent indentation
# age = 20
# if age >= 18:
#   print("Adult")    # 2 spaces
#     print("Can vote")  # 4 spaces - IndentationError!

# ✅ CORRECT: Consistent indentation (4 spaces)
age = 20
if age >= 18:
    print("Adult")
    print("Can vote")
````

### Error 2: Missing Colon

````python
# ❌ WRONG: Missing colon
# if age >= 18
#     print("Adult")  # SyntaxError

# ✅ CORRECT:
age = 20
if age >= 18:
    print("Adult")
````

### Error 3: Using = instead of ==

````python
# ❌ WRONG: Assignment instead of comparison
# age = 20
# if age = 18:  # SyntaxError
#     print("Eighteen")

# ✅ CORRECT:
age = 20
if age == 18:
    print("Eighteen")
else:
    print("Not eighteen")
````

### Application 1: Simple Calculator

````python
# Simple calculator with menu
print("=== Simple Calculator ===")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = int(input("\nEnter choice (1-4): "))
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == 1:
    result = num1 + num2
    print(f"\n{num1} + {num2} = {result}")
elif choice == 2:
    result = num1 - num2
    print(f"\n{num1} - {num2} = {result}")
elif choice == 3:
    result = num1 * num2
    print(f"\n{num1} × {num2} = {result}")
elif choice == 4:
    if num2 != 0:
        result = num1 / num2
        print(f"\n{num1} ÷ {num2} = {result}")
    else:
        print("\nError: Division by zero!")
else:
    print("\nInvalid choice!")
````

### Application 2: Tax Calculator

````python
# Income tax calculator (simplified Indian tax slabs)
income = float(input("Enter annual income (₹): "))

if income <= 250000:
    tax = 0
    tax_rate = 0
elif income <= 500000:
    tax = (income - 250000) * 0.05
    tax_rate = 5
elif income <= 1000000:
    tax = 12500 + (income - 500000) * 0.20
    tax_rate = 20
else:
    tax = 112500 + (income - 1000000) * 0.30
    tax_rate = 30

net_income = income - tax

print("\n=== Tax Calculation ===")
print(f"Gross Income: ₹{income:,.2f}")
print(f"Tax Rate: {tax_rate}%")
print(f"Tax Amount: ₹{tax:,.2f}")
print(f"Net Income: ₹{net_income:,.2f}")
````

### Application 3: Quadratic Equation Solver

````python
# Solve ax² + bx + c = 0
import math

print("Quadratic Equation Solver: ax² + bx + c = 0")
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

if a == 0:
    print("\nNot a quadratic equation (a cannot be 0)")
else:
    discriminant = b**2 - 4*a*c

    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        print(f"\nTwo real roots:")
        print(f"Root 1 = {root1:.2f}")
        print(f"Root 2 = {root2:.2f}")
    elif discriminant == 0:
        root = -b / (2*a)
        print(f"\nOne real root: {root:.2f}")
    else:
        real_part = -b / (2*a)
        imag_part = math.sqrt(abs(discriminant)) / (2*a)
        print(f"\nComplex roots:")
        print(f"Root 1 = {real_part:.2f} + {imag_part:.2f}i")
        print(f"Root 2 = {real_part:.2f} - {imag_part:.2f}i")
````
