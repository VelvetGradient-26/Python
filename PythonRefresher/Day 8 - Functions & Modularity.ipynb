## 1. Why Functions?

Functions help you:
- Avoid repeating code (DRY [Dont' Repeat Yourself]principle)
- Organize logic into small, testable units
- Improve readability and reuse

## 3. Return Values


````python
def add_and_subtract(a, b):
    result = a + b
    res2 = a - b
    return [result,res2]

sum_val = add_and_subtract(3,5)
print('3 + 5 =', sum_val)
````

## 4. Default and Keyword Arguments


````python
num = 5
if num>1:
  value = 10
  print("Internal / Local to if:",value)
  print(num)

value = 20
print("Global variable:",value)
````

````python
def power(base = 2, exponent=2):
    return base ** exponent

print(power(5,2))

print(power(5))

print(power())
````

## 5. Variable Scope

- Variables inside functions are **local**
- Global variables exist outside functions


````python
x = 10  # global

def modify():
    x = 20  # local
    print('Inside:', x)

modify()
print('Outside:', x)
````

## 6. Utility Functions
Let's build small reusable functions.


````python
def bmi(weight, height):
    return weight / (height ** 2)

def grade_from_percentage(p):
    if p >= 90:
      return 'A'
    elif p >= 80:
      return 'B'
    elif p >= 70:
      return 'C'
    elif p >= 60:
      return 'D'
    elif p >= 40:
      return 'E'
    else:
      return 'F'

print('BMI(70kg,1.75m)=', bmi(70,1.75))
print('Grade for 85% =', grade_from_percentage(85))
````

## 8. Day 8 Summary

You learned how to design, call, and reuse functions, and started thinking in terms of modular building blocks.


````python
def is_prime(n):
  flag = True

  if n<2:
    flag = False

  if n==2:
    flag = True

  for m in range(2, int(n**(1/2))+1):
    if n % m == 0:
      flag = False
      break

  return flag

var = is_prime(4)
if var:
  print("Number is PRIME")
else:
  print("Number is NOT PRIME")
````
