## 1. Understanding Data Types

Every value in Python has a **data type** that determines:
- What operations can be performed on it
- How much memory it uses
- How Python interprets the value

### Main Data Types:
| Type | Description | Example |
|------|-------------|----------|
| `int` | Whole numbers | `42`, `-10`, `0` |
| `float` | Decimal numbers | `3.14`, `-0.5`, `2.0` |
| `str` | Text/strings | `"Hello"`, `'Python'` |
| `bool` | True/False | `True`, `False` |

### Integer (int)
Whole numbers without decimal points

```python
# Integer examples
age = 25
year = 2026
temperature = -5
zero = 0

print("Age:", age)
print("Year:", year)
print("Temperature:", temperature)

# Check the type
print("\nType of age:", type(age))
```

### Float (float)
Numbers with decimal points

```python
# Float examples
pi = 3.14159
height = 5.9
price = 99.99
negative = -3.5

print("Pi:", pi)
print("Height:", height)
print("Price:", price)

# Check the type
print("\nType of pi:", type(pi))

# Even whole numbers with decimal point are floats
whole_as_float = 10.0
print("Type of 10.0:", type(whole_as_float))
```

### String (str)
Text data enclosed in quotes

```python
# String examples
name = "Alice"
city = 'Mumbai'
message = "Hello, World!"
number_as_string = "123"  # This is text, not a number!

print("Name:", name)
print("City:", city)

# Check types
print("\nType of name:", type(name))
print("Type of '123':", type(number_as_string))
```

### Boolean (bool)
True or False values - used for logic and conditions

```python
# Boolean examples
is_student = True
has_license = False
is_raining = True

print("Is student:", is_student)
print("Has license:", has_license)

# Check type
print("\nType of True:", type(is_student))

# Note: Boolean values are capitalized!
# True and False (not true/false)
```

## 2. Type Conversion (Type Casting)

Converting from one data type to another.

### Conversion Functions:
- `int()` - Convert to integer
- `float()` - Convert to float
- `str()` - Convert to string
- `bool()` - Convert to boolean

### Converting to Integer: int()

```python
# Float to int (removes decimal part)
price = 99.99
price_int = int(price)
print(f"Original: {price}, Converted: {price_int}")

# String to int
age_string = "25"
age_number = int(age_string)
print(f"String: '{age_string}', Integer: {age_number}")
print(f"Can now do math: {age_number + 5}")

# Boolean to int
print(f"\nTrue as int: {int(True)}")   # 1
print(f"False as int: {int(False)}")  # 0
```

### Converting to Float: float()

```python
# Int to float
age = 25
age_float = float(age)
print(f"Integer: {age}, Float: {age_float}")

# String to float
price_string = "49.99"
price = float(price_string)
print(f"String: '{price_string}', Float: {price}")

# Boolean to float
print(f"\nTrue as float: {float(True)}")   # 1.0
print(f"False as float: {float(False)}")  # 0.0
```

### Converting to String: str()

```python
# Number to string
age = 25
age_string = str(age)
print(f"Number: {age}, String: '{age_string}'")
print(f"Type: {type(age)} -> {type(age_string)}")

# Now can concatenate with strings
message = "I am " + str(age) + " years old"
print(message)

# Float to string
price = 99.99
print("Price: $" + str(price))
```

### ⚠️ Common Conversion Errors

```python
# This will work
number = int("123")
print(f"'123' converted to int: {number}")

# This will cause an ERROR (uncomment to see):
# number = int("123.45")  # ValueError: invalid literal

# Solution: Convert to float first, then int
number = int(float("123.45"))
print(f"'123.45' -> float -> int: {number}")

# This will also ERROR (uncomment to see):
# number = int("hello")  # ValueError: invalid literal
```

## 3. Arithmetic Operators

Python supports all standard mathematical operations.

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `10 + 5` | `15` |
| `-` | Subtraction | `10 - 5` | `5` |
| `*` | Multiplication | `10 * 5` | `50` |
| `/` | Division (float) | `10 / 3` | `3.333...` |
| `//` | Floor Division | `10 // 3` | `3` |
| `%` | Modulo (remainder) | `10 % 3` | `1` |
| `**` | Exponentiation | `2 ** 3` | `8` |

### Addition, Subtraction, Multiplication

```python
# Basic arithmetic
a = 10
b = 3

print(f"a = {a}, b = {b}")
print(f"\nAddition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
```

### Division Operations

```python
a = 10
b = 3

# Regular division (always returns float)
result = a / b
print(f"Division: {a} / {b} = {result}")
print(f"Type: {type(result)}")

# Floor division (returns integer part)
result_floor = a // b
print(f"\nFloor Division: {a} // {b} = {result_floor}")

# Modulo (remainder)
remainder = a % b
print(f"Modulo: {a} % {b} = {remainder}")

# Verification: 10 = 3 * 3 + 1
print(f"\nVerification: {b} * {result_floor} + {remainder} = {b * result_floor + remainder}")
```

### Exponentiation (Power)

```python
# Power operator **
print(f"2 ** 3 = {2 ** 3}")      # 2^3 = 8
print(f"5 ** 2 = {5 ** 2}")      # 5^2 = 25
print(f"10 ** 0 = {10 ** 0}")    # Anything^0 = 1

# Square root (power of 0.5)
number = 16
square_root = number ** 0.5
print(f"\nSquare root of {number} = {square_root}")

# Cube root
cube_root = 27 ** (1/3)
print(f"Cube root of 27 = {cube_root}")
```

## 4. Operator Precedence (Order of Operations)

When multiple operators are used, Python follows PEMDAS:
1. **P**arentheses `()`
2. **E**xponentiation `**`
3. **M**ultiplication/Division/Modulo `*`, `/`, `//`, `%` (left to right)
4. **A**ddition/Subtraction `+`, `-` (left to right)

```python
# Without parentheses
result = 2 + 3 * 4
print(f"2 + 3 * 4 = {result}")  # Multiplication first: 2 + 12 = 14

# With parentheses
result = (2 + 3) * 4
print(f"(2 + 3) * 4 = {result}")  # Parentheses first: 5 * 4 = 20

# Complex expression
result = 10 + 5 * 2 ** 2
print(f"\n10 + 5 * 2 ** 2 = {result}")
print("Steps: 2**2=4, then 5*4=20, then 10+20=30")

# Make it clear with parentheses
result = 10 + (5 * (2 ** 2))
print(f"With parentheses: {result}")
```

## 5. Practical Applications

### Application 1: Temperature Converter
Formula: F = (C × 9/5) + 32

```python
# Celsius to Fahrenheit
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C = {fahrenheit}°F")

# Fahrenheit to Celsius
fahrenheit = float(input("\nEnter temperature in Fahrenheit: "))
celsius = (fahrenheit - 32) * 5/9
print(f"{fahrenheit}°F = {celsius:.2f}°C")
```

### Application 2: BMI Calculator
Formula: BMI = weight(kg) / (height(m))²

```python
# BMI Calculator
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

bmi = weight / (height ** 2)
print(f"\nYour BMI: {bmi:.2f}")

# BMI Categories (preview for later)
print("\nBMI Categories:")
print("Below 18.5: Underweight")
print("18.5-24.9: Normal")
print("25-29.9: Overweight")
print("30+: Obese")
```

### Application 3: Simple & Compound Interest
Simple Interest: SI = (P × R × T) / 100  
Compound Interest: CI = P × (1 + R/100)^T - P

```python
# Interest Calculator
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest (%): "))
time = float(input("Enter time (years): "))

# Simple Interest
simple_interest = (principal * rate * time) / 100
print(f"\nSimple Interest: ₹{simple_interest:.2f}")
print(f"Total Amount: ₹{principal + simple_interest:.2f}")

# Compound Interest
amount = principal * ((1 + rate/100) ** time)
compound_interest = amount - principal
print(f"\nCompound Interest: ₹{compound_interest:.2f}")
print(f"Total Amount: ₹{amount:.2f}")
```

### Application 4: Currency Converter

```python
# Currency Converter (INR to USD, EUR, GBP)
inr_amount = float(input("Enter amount in INR: "))

# Exchange rates (example rates)
usd_rate = 0.012  # 1 INR = 0.012 USD
eur_rate = 0.011  # 1 INR = 0.011 EUR
gbp_rate = 0.009  # 1 INR = 0.009 GBP

print(f"\n₹{inr_amount} INR is equal to:")
print(f"${inr_amount * usd_rate:.2f} USD")
print(f"€{inr_amount * eur_rate:.2f} EUR")
print(f"£{inr_amount * gbp_rate:.2f} GBP")
```

### Application 5: Restaurant Bill Calculator

```python
# Restaurant Bill with Tax and Tip
bill_amount = float(input("Enter bill amount: ₹"))
tax_rate = float(input("Enter tax rate (%): "))
tip_rate = float(input("Enter tip rate (%): "))

# Calculate
tax = bill_amount * (tax_rate / 100)
tip = bill_amount * (tip_rate / 100)
total = bill_amount + tax + tip

# Display receipt
print("\n" + "="*40)
print("RESTAURANT BILL")
print("="*40)
print(f"Subtotal:\t₹{bill_amount:.2f}")
print(f"Tax ({tax_rate}%):\t₹{tax:.2f}")
print(f"Tip ({tip_rate}%):\t₹{tip:.2f}")
print("-" * 40)
print(f"TOTAL:\t\t₹{total:.2f}")
print("="*40)
```
