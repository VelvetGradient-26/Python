## 5. Escape Characters

Special characters that change how text is displayed.

| Escape Sequence | Description | Example |
|----------------|-------------|----------|
| `\n` | New line | Moves to next line |
| `\t` | Tab | Adds horizontal space |
| `\\` | Backslash | Displays \ character |
| `\'` | Single quote | Displays ' inside single quotes |
| `\"` | Double quote | Displays " inside double quotes |

````python
print("20\")
````

````python
# New line character \n
print("Line 1\nLine 2\nLine 3")

print()  # Empty line for spacing

# Tab character \t
print("Name\tAge\tCity")
print("Alice\t25\tNYC")
print("Bob\t30\tLA")

print()  # Empty line for spacing

# Combining escape characters
print("First Line\n\tIndented Second Line")
````

## 6. Variables and Strings

### What is a Variable?
- A container that stores data
- Has a name (identifier) and a value
- Can be changed later in the program

### Variable Naming Rules:
- Must start with letter or underscore
- Can contain letters, numbers, underscores
- Case-sensitive (`name` ≠ `Name`)
- Cannot use Python keywords (like `print`, `if`, `for`, etc.)

````python
# Creating variables
name = "Alice"
age = 25
city = "Mumbai"

# Printing variables
print(name)
print(age)
print(city)

# Printing with labels
print("Name:", name)
print("Age:", age)
print("City:", city)
````

## 7. String Concatenation

**Concatenation** = joining strings together using the `+` operator

### Important:
- Can only concatenate strings with strings
- Numbers must be converted to strings first using `str()`

````python
# String concatenation
first_name = "John"
last_name = "Doe"

full_name = first_name + " " + last_name
print(full_name)

# Concatenating with text
#greeting = "Hello, " + first_name + "!"
#print(greeting)
````

````python
# Concatenating strings with numbers (requires conversion)
name = "Alice"
age = 25

# This will cause an ERROR (uncomment to see):
# print("Name: " + name + ", Age: " + age)

# CORRECT way: convert number to string
print("Name: " + name + ", Age: " + str(age))
````

## 8. F-Strings (Formatted String Literals)

**F-strings** are the modern, cleaner way to format strings in Python!

### Syntax:
- Put `f` before the opening quote
- Use `{}` to insert variables

### Advantages:
- More readable
- No need to convert numbers to strings
- Can include expressions inside `{}`

````python
# F-strings examples
name = "Alice"
age = 25
city = "Mumbai"

# Simple f-string
print(f"Name: {name}")
print(f"Age: {age}")

# Multiple variables
print(f"{name} is {age} years old and lives in {city}.")

# With expressions
print(f"Next year, {name} will be {age + 1} years old.")
````

## 9. Getting User Input

The `input()` function allows your program to interact with users.

### How it works:
1. Program displays a prompt (optional)
2. Waits for user to type something
3. User presses Enter
4. Input is stored as a **string**

### Important:
- `input()` always returns a **string** (text)
- Even if user types a number, it's stored as text

````python
# Basic input
name = input("What is your name? ")
print("Hello, " + name + "!")
````

````python
# Input with f-strings
name = input("Enter your name: ")
age = input("Enter your age: ")  # This is a STRING, not a number!

print(f"Hello {name}! You are {age} years old.")
````

## 10. Common Syntax Errors

Learning to identify and fix errors is a crucial programming skill!

### Types of Errors:
1. **Syntax Errors**: Code doesn't follow Python rules
2. **Runtime Errors**: Code runs but crashes
3. **Logic Errors**: Code runs but gives wrong results

### Error 1: Missing Quotes
```python
# ❌ WRONG:
print(Hello)  # NameError: name 'Hello' is not defined

# ✅ CORRECT:
print("Hello")
```

### Error 2: Mismatched Quotes
```python
# ❌ WRONG:
print('Hello")  # SyntaxError

# ✅ CORRECT:
print('Hello')  # or print("Hello")
```

### Error 3: Missing Parentheses
```python
# ❌ WRONG:
print "Hello"  # SyntaxError (Python 3)

# ✅ CORRECT:
print("Hello")
```

### Error 4: Extra Parentheses
```python
# ❌ WRONG:
print("Hello"))  # SyntaxError

# ✅ CORRECT:
print("Hello")
```

### Error 5: Concatenating String with Number
```python
# ❌ WRONG:
age = 25
print("Age: " + age)  # TypeError

# ✅ CORRECT:
print("Age: " + str(age))  # Convert to string
# OR use f-string:
print(f"Age: {age}")
```

## 14. Tips and Best Practices

### Good Coding Habits:
1. **Use meaningful variable names**
   - ✅ `student_name` or `studentName`
   - ❌ `x` or `var1`

2. **Add comments to explain complex logic**
   - But don't over-comment obvious things

3. **Use f-strings for formatting**
   - More readable than concatenation
   - Automatically handles type conversion

4. **Add blank lines for readability**
   - Separate logical sections of code

5. **Test your code frequently**
   - Run small portions as you write
   - Fix errors as they appear

### Debugging Tips:
- **Read error messages carefully** - they tell you what's wrong
- **Check syntax first** - quotes, parentheses, colons
- **Use print() to debug** - see what values variables hold
- **Google the error** - someone has likely solved it before
- **Ask for help** - don't struggle alone for too long
