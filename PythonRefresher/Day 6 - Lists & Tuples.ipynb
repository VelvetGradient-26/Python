## 2. Indexing and Slicing

Lists are **ordered**, so each element has an index. Python uses **0-based indexing**.

### Indexing:
- `lst[0]` → first element
- `lst[-1]` → last element

### Slicing:
- `lst[start:stop]` (stop not included)
- `lst[start:stop:step]`


````python
lst = [10, 20, 30, 40, 50]
print(lst[0])
print(lst[-1])
````

````python
data = [11, 22, 33, 44, 55, 66, 77, 88, 99]

print('data:', data)
print(data[::])
# Indexing
print('First element:', data[0])
print('Third element:', data[2])
print('Last element:', data[-1])

# Slicing
print("First 5 elements:", data[:5])
print('Elements 1 to 3:', data[1:4])   # indices 1,2,3
print('First three:', data[:3])
print('From index 2 to end:', data[2:])
print('Every second element:', data[::2])
print('Reversed:', data[::-1])

print("Elements of index -1 to -5:", data[-1:-6:-1])
````

## 3. Common List Methods

Useful methods for modifying lists in-place:

- `append(x)` → add item to end
- `extend(iterable)` → add all items from another iterable
- `insert(i, x)` → insert at position i
- `remove(x)` → remove first occurrence of x
- `pop([i])` → remove and return element at i (last by default)
- `sort()` → sort in-place
- `reverse()` → reverse in-place


````python
fruits = ['apple', 'banana', 'orange']
print('Start:', fruits)

fruits.append('mango')
print('After append:', fruits)

fruits.insert(1, 'kiwi')
print('After insert at 1:', fruits)

fruits.remove('banana')
print('After remove banana:', fruits)

last = fruits.pop()
print('After pop:', fruits, '| popped:', last)

fruits.sort()
print('Sorted:', fruits)

fruits.reverse()
print('Reversed:', fruits)
````

## 4. Built-in Functions with Lists

Common built-ins for numeric lists:

- `len(lst)` → number of elements
- `sum(lst)` → sum of elements
- `min(lst)`, `max(lst)`
- `sorted(lst)` → returns a new sorted list (does not modify original)


````python
scores = [88, 92, 75, 63, 95, 80]
print('Scores:', scores)

print('Count:', len(scores))
print('Sum:', sum(scores))
print('Average:', sum(scores) / len(scores))
print('Min:', min(scores))
print('Max:', max(scores))
print('Sorted ascending:', sorted(scores))

print('Original unchanged:', scores)
````

## 5. Iterating Over Lists

You can loop through list elements directly or with indices.


````python
"""
Direct iteration
i = 'apple', prints apple
i = 'banana', prints banana
i = 'mango', prints mango


With index (range)
i = 0, prints fruits[0] = apple
i = 1, prints fruits[1] = banana
i = 2, prints fruits[2] = mango
"""
````

````python
fruits = ['apple', 'banana', 'mango']

print('Direct iteration:')
for i in fruits:
    print(i)

print('With index (range):')
for i in range(len(fruits)):
    print(i, '->', fruits[i])
````

## 6. List Comprehensions (Intro)

A **list comprehension** is a concise way to create lists.

```python
squares = [x**2 for x in range(1, 6)]
```


````python
# Traditional way
squares = []
for x in range(1, 6):
    squares.append(x**2)
print('Traditional:', squares)

# List comprehension
squares2 = [x**2 for x in range(1, 6)]
print('Comprehension:', squares2)

# Filtered comprehension
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print('Even squares:', even_squares)
````

## 7. Tuples

A **tuple** is an ordered, **immutable** collection. Once created, it cannot be changed.

### Why Tuples?
- Protect data from modification
- Slightly more memory-efficient
- Often used for fixed-size records (coordinates, RGB colors, etc.)


````python
point = (10, 20)
color = (255, 0, 0)  # RGB
print('Point:', point)
print('Color:', color)

print('First coord:', point[0])
print('Second coord:', point[1])

# Tuple unpacking
x, y = point
print('x =', x, ', y =', y)
````

## 8. Lists vs Tuples

| Feature | List | Tuple |
|--------|------|-------|
| Syntax | `[]` | `()` |
| Mutability | Mutable | Immutable |
| Typical use | Changing collections | Fixed records |


````python
name = input("Enter the name: ")
vowels = ['a','e','i','o','u','A','E','I',"O","U"]
count = 0

for char in name:
  if char in vowels:
    count += 1
print(count)
````
