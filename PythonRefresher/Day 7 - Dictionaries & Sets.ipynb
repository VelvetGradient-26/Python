## 1. Dictionary Basics

A **dictionary** stores data as **key-value pairs**.

### Syntax:
```python
student = {'name': 'Alice', 'age': 20}
```


````python
student = {'name': 'Alice', 'age': 20, 'marks': 88}
print('Student:', student)

# Access values
print('Name:', student['name']) #dictionary_name[key]
print('Age:', student['age'])

# Add/update
student['city'] = 'Jammu'
student['age'] = 21
print('Updated:', student)
````

## 2. Accessing Values Safely

Using `[]` raises `KeyError` if key is missing. `get()` returns `None` or a default.


````python
print(student.get('name'))
print(student.get('phone'))           # None
print(student.get('phone', 'No Key named as phone found here'))  # default value
````

## 3. Iterating Over Dictionaries


````python
print('Keys:')
for k in student.keys():
    print(k)

print('Values:')
for v in student.values(): # dictionary_name.values()
    print(v)

print('Items:')
for k, v in student.items(): #dictionary_name.items()
    print(k, ':', v)
````

## 4. Useful Dictionary Methods

- `keys()`, `values()`, `items()`
- `update()` – merge another dict
- `pop(key)` – remove and return value
- `popitem()` – remove last inserted pair
- `clear()` – remove all items


````python
config = {'host': 'localhost', 'port': 5432}

print('Config:', config)
config.update({'user': 'admin', 'IP': 190.234})
print('After update:', config)

p = config.pop('port')
print('Popped port:', p)
print('After pop:', config)
````

## 5. Nested Dictionaries

Dictionaries can contain other dictionaries (similar to JSON).


````python
students = {
    101: {'name': 'Amit', 'marks': 85},
    102: {'name': 'Sara', 'marks': 92}
}

print(students)
print(students[101]['name'])

print(students.get(101).get('name'))
````

## 6. Frequency Counter (Classic DS Use-Case)


````python
"""
freq = {}
w = "apple", if w in freq --> False, freq['apple'] = 1
freq = {'apple': 1}
w = "banana", if w in freq --> False, freq['banana'] = 1
freq = {'apple': 1, 'banana': 1}
w = "apple", if w in freq --> True, freq['apple'] = 2
..
..
..
..

"""
````

````python
text = 'apple banana apple orange banana apple'
words = text.split()
print(words)
freq = {} # Key : Value would be the Fruit_Name:How many times it has appeared in the list
for w in words:
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1
print(freq)
````

## 7. Sets

A **set** is an unordered collection of unique elements.


````python
nums = [1,2,2,3,3,3,4]
s = set(nums)
print('Original:', nums)
print('As set:', s)
s.add(5)
print('After add:', s)
set1  = {6,7,7,10,24,53,23,64,22}
print(set1)
````

## 8. Set Operations


````python
a = {1,2,3,4}
b = {3,4,5,6}
print('a:', a)
print('b:', b)

print('Union:', a | b)
print('Intersection:', a & b)
print('Difference a-b:', a - b)
print('Symmetric diff:', a ^ b)
````
