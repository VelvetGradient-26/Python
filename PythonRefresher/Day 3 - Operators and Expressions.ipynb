## 1. Comparison Operators

Comparison operators compare two values and return **True** or **False**.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `<` | Less than | `3 < 5` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<=` | Less than or equal | `5 <= 5` | `True` |
| `>=` | Greater than or equal | `5 >= 3` | `True` |

### Equal To (==) and Not Equal To (!=)

```python
# Equal to (==)
x = 10
y = 20

print(f"x = {x}, y = {y}")
print(f"x == y: {x == y}")   # False
print(f"x == 10: {x == 10}")  # True

# Not equal to (!=)
print(f"\nx != y: {x != y}")   # True
print(f"x != 10: {x != 10}")  # False

# Important: == vs =
# = is assignment
# == is comparison
age = 25      # Assignment
a = age == 25  # Comparison
print(f"\na: {a}")
```

### Less Than (<) and Greater Than (>)

```python
# Comparison operators
a = 15
b = 20

print(f"a = {a}, b = {b}")
print(f"\na < b: {a < b}")    # True
print(f"a > b: {a > b}")      # False
print(f"a <= 15: {a <= 15}")  # True
print(f"b >= 25: {b >= 25}")  # False

# Comparing with equal value
x = 10
print(f"\nx < 10: {x < 10}")   # False
print(f"x <= 10: {x <= 10}")   # True
```

### Comparing Different Data Types

```python
# Numbers: int and float can be compared
print(f"10 == 10.0: {10 == 10.0}")  # True
print(f"5 < 5.1: {5 < 5.1}")        # True

# Strings: lexicographic (alphabetical) comparison
print(f"\n'apple' < 'banana': {'apple' < 'banana'}")  # True
print(f"'ABC' < 'abc': {'ABC' < 'abc'}")  # True (uppercase < lowercase)

# String and number comparison
print("10" == 10)  # False - different types
print(f"'10' == str(10): {'10' == str(10)}")  # True
```

## 2. Logical Operators

Logical operators combine boolean values or expressions.

| Operator | Description | Example |
|----------|-------------|----------|
| `and` | Both must be True | `True and False` → `False` |
| `or` | At least one must be True | `True or False` → `True` |
| `not` | Reverses the boolean | `not True` → `False` |

### AND Operator
Returns True only if **both** conditions are True

```python
# AND operator truth table
print("AND Truth Table:")
print(f"True and True: {True and True}")
print(f"True and False: {True and False}")
print(f"False and True: {False and True}")
print(f"False and False: {False and False}")

# Practical example
age = 25
has_license = True

can_drive = age >= 18 and has_license
print(f"\nCan drive: {can_drive}")
```

### OR Operator
Returns True if **at least one** condition is True

```python
# OR operator truth table
print("OR Truth Table:")
print(f"True or True: {True or True}")
print(f"True or False: {True or False}")
print(f"False or True: {False or True}")
print(f"False or False: {False or False}")

# Practical example
is_weekend = False
is_holiday = True

can_relax = is_weekend or is_holiday
print(f"\nCan relax: {can_relax}")
```

### NOT Operator
Reverses the boolean value

```python
# NOT operator
print("NOT Truth Table:")
print(f"not True: {not True}")
print(f"not False: {not False}")

# Practical example
is_raining = False
can_play_outside = not is_raining
print(f"\nCan play outside: {can_play_outside}")

# Double negation
print(f"not not True: {not not True}")
```

### Combining Logical Operators

```python
# Complex expressions
age = 25
income = 50000
has_credit_card = True
credit_score = 750

# Loan eligibility
is_eligible = (age >= 21 and age <= 60) and (income >= 30000) and (credit_score >= 700)
print(f"Loan eligible: {is_eligible}")

# Alternative payment methods
can_pay = has_credit_card or income > 40000
print(f"Can make payment: {can_pay}")

# Complex condition
special_case = (age < 25 or age > 60) and not has_credit_card
print(f"Special case: {special_case}")
```

## 3. Operator Precedence with Logical Operators

### Order of Evaluation:
1. Arithmetic operators (`**`, `*`, `/`, `+`, `-`)
2. Comparison operators (`<`, `>`, `==`, `!=`, etc.)
3. `not`
4. `and`
5. `or`

```python
# Precedence examples
x = 10
y = 20
z = 30

# Without parentheses
result = x < y and y < z
print(f"x < y and y < z: {result}")

# Mixing operators
result = x + 5 > y or z - 10 < y
print(f"x + 5 > y or z - 10 < y: {result}")

# NOT has highest precedence among logical
result = not x > y and z > y
print(f"not x > y and z > y: {result}")

# Use parentheses for clarity
result = (not x > y) and (z > y)
print(f"(not x > y) and (z > y): {result}")
```

### Short-Circuit Evaluation

Python stops evaluating as soon as result is determined:
- `and`: If first is False, doesn't check second
- `or`: If first is True, doesn't check second

```python
# Short-circuit with AND
x = 5
result = (x > 10) and (x / 0 > 1)  # Won't error! x > 10 is False, so stops
print(f"Result: {result}")

# Short-circuit with OR
result = (x < 10) or (x / 0 > 1)  # Won't error! x < 10 is True, so stops
print(f"Result: {result}")

# This would error (uncomment to see):
# result = (x < 10) and (x / 0 > 1)  # First is True, so checks second → Error!
```

## 4. Assignment Operators

Shortcuts for common operations.

| Operator | Example | Equivalent |
|----------|---------|-------------|
| `=` | `x = 5` | Assign 5 to x |
| `+=` | `x += 3` | `x = x + 3` |
| `-=` | `x -= 3` | `x = x - 3` |
| `*=` | `x *= 3` | `x = x * 3` |
| `/=` | `x /= 3` | `x = x / 3` |
| `//=` | `x //= 3` | `x = x // 3` |
| `%=` | `x %= 3` | `x = x % 3` |
| `**=` | `x **= 3` | `x = x ** 3` |

```python
# Addition assignment
score = 10
print(f"Initial score: {score}")

score += 5  # Same as: score = score + 5
print(f"After += 5: {score}")

score -= 3  # Same as: score = score - 3
print(f"After -= 3: {score}")

score *= 2  # Same as: score = score * 2
print(f"After *= 2: {score}")

score //= 4  # Same as: score = score // 4
print(f"After //= 4: {score}")
```

### Practical Use Cases

```python
# Counter
count = 0
count += 1  # Increment
count += 1
count += 1
print(f"Count: {count}")

# Running total
total = 0
total += 100
total += 250
total += 75
print(f"Total: {total}")

# Multiplier
value = 5
value *= 10  # Quick multiplication
print(f"Value: {value}")
```

## 5. Membership Operators

Check if value exists in a sequence.

| Operator | Description |
|----------|-------------|
| `in` | Returns True if value is in sequence |
| `not in` | Returns True if value is NOT in sequence |

```python
# in operator with strings
message = "Hello, Python!"
print(f"'Python' in message: {'Python' in message}")
print(f"'Java' in message: {'Java' in message}")

# Check substring
email = "student@iitjammu.ac.in"
is_iit = "iitjammu" in email
print(f"\nIs IIT email: {is_iit}")

# not in operator
text = "Python Programming"
print(f"\n'Java' not in text: {'Java' not in text}")
```

## 6. Identity Operators

Check if two variables point to same object in memory.

| Operator | Description |
|----------|-------------|
| `is` | Returns True if same object |
| `is not` | Returns True if different objects |

```python
# Identity vs Equality
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f"a == b: {a == b}")  # True (same values)
print(f"a is b: {a is b}")  # False (different objects)
print(f"a is c: {a is c}")  # True (same object)

# With numbers (small integers cached)
x = 10
y = 10
print(f"\nx is y: {x is y}")  # True for small integers

# Checking None
value = None
print(f"\nvalue is None: {value is None}")  # Correct way
```

## 7. Practical Applications

### Application 1: Eligibility Checker

```python
# University admission eligibility
print("=== University Admission Checker ===")
age = int(input("Enter age: "))
marks = float(input("Enter percentage: "))
has_diploma = input("Have diploma? (yes/no): ").lower() == "yes"

# Eligibility criteria
age_eligible = age >= 17 and age <= 25
marks_eligible = marks >= 75
overall_eligible = age_eligible and (marks_eligible or has_diploma)

print(f"\nAge eligible: {age_eligible}")
print(f"Marks eligible: {marks_eligible}")
print(f"Overall eligible: {overall_eligible}")
```

### Application 2: Password Strength Checker

```python
# Basic password strength checker
password = input("Enter password: ")

# Check criteria
length_ok = len(password) >= 8
has_upper = any(c.isupper() for c in password)
has_lower = any(c.islower() for c in password)
has_digit = any(c.isdigit() for c in password)

# Simple checks for now
print(f"\nLength >= 8: {length_ok}")
print(f"Has uppercase: {has_upper}")
print(f"Has lowercase: {has_lower}")
print(f"Has digit: {has_digit}")

is_strong = length_ok and has_upper and has_lower and has_digit
print(f"\nStrong password: {is_strong}")
```

### Application 3: Number Validator

```python
# Number properties checker
num = int(input("Enter a number: "))

# Check properties
is_positive = num > 0
is_even = num % 2 == 0
is_divisible_by_5 = num % 5 == 0
is_two_digit = num >= 10 and num <= 99
is_in_range = num >= 1 and num <= 100

print(f"\nNumber: {num}")
print(f"Positive: {is_positive}")
print(f"Even: {is_even}")
print(f"Divisible by 5: {is_divisible_by_5}")
print(f"Two digit: {is_two_digit}")
print(f"In range 1-100: {is_in_range}")
```
