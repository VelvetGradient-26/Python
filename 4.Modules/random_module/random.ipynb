```python
import random
```

```python
# print a float in the range from 0 - 1
random.random()
```

```python
# print a random int for a specified range
random.randint(1,100)
```

```python
# randrange works similar to randint, but with a step count
random.randrange(1,10,2)
```

```python
# pick a random element from an iterable
lst = [x for x in range(1,26)]
random.choice(lst)

# you add the size of the output too, using the choices() method
random.choices(lst, k=5)
```

```python
# randomly shuffle the list
random.shuffle(lst)
print(lst)
```

```python
# use the seed function to anchor the randomness for once
random.seed(10)

random_lst = [random.randint(1,100) for x in range(1,26)]
print(random_lst)
```
