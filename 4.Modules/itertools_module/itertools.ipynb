```python
import itertools as it
```

```python
lst = [1,2,3,4,5,6]
lst
```

```python
# permutation function
pers = it.permutations(lst, r = 2)
count = 0

for per in pers: 
    count += 1
    print(per)

print(count)
```

```python
# use the combination function 

combs = it.combinations(lst, r = 2)
count = 0

for comb in combs: 
    count += 1
    print(comb)

print("Count: ", count)
```

```python
# cartesian product
lst = [[1,2,3], ['a', 'b']]

cart_lst = it.product(*lst)
for cart in cart_lst: 
    print(cart)
```
