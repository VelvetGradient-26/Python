````python
# Basic class with a constructor

class Apple:
    def __init__(self, color, flavor):
        self.color = color
        self.flavor = flavor

    # controls what is printed when the print function is used on the object
    def __str__(self):
        return f"An apple which is {self.color} in color and {self.flavor} in taste"

honeycrisp = Apple('Red', 'Sweet')
fuji = Apple("Red", "Tart")

print(honeycrisp)
print(fuji)
````

````python
# Rectangle class
class Rectangle: 
    # class variable
    count = 0
    # self is the reference to the current object
    def __init__(self, length, breadth): 
        # Instance variables
        self.length = length 
        self.breadth = breadth

        # class variable
        Rectangle.count += 1

    # instance methods
    def perimeter(self):
        return 2 *(self.length + self.breadth)
    
    def area(self): 
        return self.length * self.breadth
    
    # class method; should use the classmethod decorator
    @classmethod
    def get_count(cls): 
        return cls.count
    
    @staticmethod
    def odd_or_even(x): 
        return "Even" if x % 2 == 0 else "Odd"

rec = Rectangle(10, 5)
print(rec.perimeter())
print(rec.area())

rect = Rectangle(10, 20)
print(rect.perimeter())
print(rect.area())

print(Rectangle.get_count())

print(rect.odd_or_even(874))
````

````python
# Property Method

class Rectangle: 
    def __init__(self, l, b): 
        self.length = l
        self.breadth = b

    @property
    def length(self): 
        return self._length
    
    @length.setter
    def length(self, l): 
        if l >= 0: 
            self._length = l
        else: 
            self._length = 1

    @property
    def breadth(self): 
        return self._breadth
    
    @breadth.setter
    def breadth(self, b): 
        if b >= 0: 
            self._breadth = b
        else: 
            self._breadth = 1

rect = Rectangle(10, 5)
print(rect.length)
print(rect.breadth)

rect.length = -9
print(rect.length)

rect.breadth = -1
print(rect.breadth)
````

````python
# INHERITANCE

# creating a cuboid class using the Rectangle class

class Cuboid(Rectangle): 
    def __init__(self, l, b, h): 
        super().__init__(l,b)
        self.height = h

    def volume(self): 
        return self.length * self.breadth * self.height
    
cuboid = Cuboid(10,5,3)
print(cuboid.volume())
````

````python
# Polymorphism - Duck Typing

class Duck: 
    def talk(self): 
        print("Duck is Talking")

    def walk(self): 
        print("Duck is walking")

class Dog: 
    def talk(self): 
        print("Dog is talking")
    
    def walk(self): 
        print("Dog is walking")

class Horse: 
    def talk(self): 
        print("Horse is talking")
    
    def run(self): 
        print("Horse is running")

def person(pet): 
    pet.talk()
    if hasattr(pet, 'walk'):
        pet.walk()

dog = Dog()
person(dog)

duck = Duck()
person(duck)

horse = Horse()
person(horse)
        
````

````python
# Polymorphism - Method Overloading (Same function name, with different: Type of parameters, number of parameters)

# method overloading is enabled by default in python for type of parameters

# works for all data types
def sum(a, b): 
    return a + b


# the above function is shadowed and this is the new 'sum' function
def sum(a, b, c): 
    return a + b + c
````

````python
# Polymorphism - Method Overriding

class Parent: 
    def show(self): 
        print("Parent Show")

    def display(self): 
        print("I am a Parent")

class Child(Parent): 
    # this method will override the show method from the parent class
    def show(self): 
        print("Child Show")

    # you can get both of them via the super() method
    def display(self): 
        super().display()
        print("I am a Child")
    

c = Child()
c.show()

print(c.display)
````

````python
# Operator level Polymorphism 

from math import pi, pow

class Circle: 
    def __init__(self, diameter):
        self.diameter = diameter
        self.radius = diameter / 2

    def circumference(self):
        return 2 * pi * self.radius
    
    def area(self): 
        return pi * pow(self.radius, 2)
    
    def __add__(self, other): 
        return self.area() + other.area()
    
circle1 = Circle(347.67)
circle2 = Circle(123.78)
        
print(f"Area of circle 1: {circle1.area():.2f}")
print(f"Area of circle 2: {circle2.area():.2f}")

print(circle1 + circle2)
````

````python
# Add The volume of two cuboids

class Cuboid(Rectangle): 
    def __init__(self, length, breadth, height): 
        super().__init__(length, breadth)
        self.height = height

    def volume(self): 
        return self.length * self.height * self.breadth
    
    def __add__(self, other): 
        return self.volume() + other.volume()
    
cuboid1 = Cuboid(10, 20, 5)
print(cuboid1.volume())

cuboid2 = Cuboid(10, 5, 3)
print(cuboid2.volume())

print(cuboid1 + cuboid2)
````

````python
# Adding two vectors

class Vector: 
    def __init__(self, x, y): 
        self.x = x
        self.y = y

    def __add__(self, other): 
        x = self.x + other.x
        y = self.y + other.y
        return (x, y)

v1 = Vector(10, 2)
v2 = Vector(2, 3)

print(v1 + v2)
````

````python
from abc import ABC, abstractmethod

class Parent(ABC): 
    def meth1(self): 
        print("This is method 1 from Parent Class")

    @abstractmethod
    def meth2(self): 
        print("This is method 2 from Parent class")


class Child(Parent): 
    def meth3(self): 
        print("This is Child Method 3 from Child Class")

    def meth2(self): 
        print("This is method 2 from Child Class (Overidden Method)")

c = Child()
c.meth1()
c.meth2()
c.meth3()
````

### **Class: The Object Blueprint** 📝

A **class** serves as the template or blueprint for creating objects. It defines the structure and behavior that its objects will have. Every class is composed of two main elements:

* **Data Members / Properties:** These are the variables that hold the state or data of an object (e.g., `color`, `name`, `age`).
* **Functions / Methods:** These define the actions an object can perform or the operations that can be applied to it (e.g., `calculateArea()`, `startEngine()`).

---

### **Pillars of OOP** 🏛️

Object-Oriented Programming is built upon four fundamental principles that help in creating scalable, manageable, and reusable code.

#### **Encapsulation**

Encapsulation is the practice of bundling an object's data (properties) and the methods that operate on that data into a single unit, the class. It restricts direct access to an object's internal state from the outside, a concept known as **data hiding**. This protects the data from accidental modification and makes the code more secure and robust.

#### **Abstraction**

Abstraction focuses on hiding complex implementation details while exposing only the essential functionalities to the user. It simplifies the interface of an object by emphasizing *what* it does rather than *how* it does it. For example, you know how to use a remote control to change the channel without needing to understand its internal circuitry.

#### **Inheritance**

Inheritance is a mechanism that allows a new class (the **child** or **subclass**) to acquire the properties and methods of an existing class (the **parent** or **superclass**). This promotes code reusability ("write once, use everywhere") and establishes a logical "is-a" relationship between classes (e.g., a `Dog` is an `Animal`).

#### **Polymorphism**

Meaning "many forms," polymorphism allows objects to be treated differently based on their data type or class. It enables a single interface (like a method name) to represent different underlying forms (implementations). A common example is a `makeSound()` method, which would produce a "bark" for a `Dog` object and a "meow" for a `Cat` object, even though the same method name is called.

### Generalization and Specialization

These are fundamental concepts in OOP that describe the relationship between classes.

  * **Generalization (Is-A relationship from child to parent):**

      * This is the process of extracting common properties (attributes) and behaviors (methods) from specialized classes and defining them in a more general (parent or base) class.
      * It represents an "is a type of" relationship. For example, a `Car` *is a type of* `Vehicle`.
      * It's achieved through **inheritance**, where a subclass inherits from a superclass.
      * The general class captures what is common, while the specialized classes define what makes them unique.
      * **Analogy:** Thinking about "vehicles" as a general category for "cars," "motorcycles," and "trucks."

  * **Specialization (Is-A relationship from parent to child):**

      * This is the process of creating more specific classes (subclasses or derived classes) from a more general class (superclass or base class).
      * The specialized class adds unique attributes and behaviors, or overrides inherited behaviors to provide specific implementations.
      * It refines or extends the general concept.
      * **Analogy:** A "SportsCar" is a specialization of a "Car," adding features like higher horsepower and specialized suspension.

In essence, **generalization moves upwards in a class hierarchy**, identifying commonalities, while **specialization moves downwards**, adding unique characteristics.

### Types of Classes

These different types of classes help manage the balance between generalization and specialization, enforcing design principles like abstraction and polymorphism.

#### 1\. Concrete Class

  * **Definition:** A concrete class is a class that can be **instantiated** (i.e., you can create direct objects/instances of it). It provides a complete implementation for all its methods.

  * **Characteristics:**

      * All its methods have an implementation (they are not abstract).
      * Can have attributes and constructors.
      * Can inherit from other classes (concrete or abstract) and can be inherited by other classes.

  * **Purpose:** To create usable objects that represent specific entities or concepts in your program.

  * **Example (Python):**

    ```python
    class Dog: # This is a concrete class
        def __init__(self, name, breed):
            self.name = name
            self.breed = breed

        def make_sound(self):
            return f"{self.name} barks!"

    my_dog = Dog("Buddy", "Golden Retriever") # You can create an instance
    print(my_dog.make_sound())
    ```

#### 2\. Interface (Concept in Python, Formal in Java/C\#)

  * **Definition:** In many languages (like Java, C\#), an interface is a *contract* that defines a set of methods that a class must implement if it "implements" that interface. It declares *what* a class should do, but not *how*. It typically contains only method signatures (abstract methods) and possibly constants.

  * **Characteristics (General OOP Concept):**

      * Cannot be instantiated directly.
      * All methods are implicitly abstract (no implementation).
      * A class can implement multiple interfaces (achieving multiple inheritance of "behavior").
      * Often used to define common behaviors across unrelated classes.

  * **Purpose:** To define a common API (Application Programming Interface) or a "role" that multiple classes can fulfill, ensuring that all implementing classes provide specific functionalities. This is a powerful tool for polymorphism.

  * **Interface in Python:** Python doesn't have a direct `interface` keyword like Java. The concept of interfaces is achieved through:

      * **Abstract Base Classes (ABCs):** Using the `abc` module, you can define abstract methods that subclasses *must* implement. This is the closest equivalent to formal interfaces.
      * **Duck Typing:** Python's dynamic nature often relies on "duck typing" ("If it walks like a duck and quacks like a duck, then it is a duck"). If an object behaves in a certain way (i.e., has the required methods), it can be treated as if it implements an "interface" even if no formal interface was declared.

  * **Example using ABC (Python's way of defining an interface-like contract):**

    ```python
    from abc import ABC, abstractmethod

    class Drawable(ABC): # Defines an interface-like contract
        @abstractmethod
        def draw(self):
            pass

    class Circle(Drawable):
        def draw(self):
            print("Drawing a circle.")

    class Square(Drawable):
        def draw(self):
            print("Drawing a square.")

    # class Triangle(ABC): # This would raise an error if not implemented draw
    #     pass

    circle = Circle()
    circle.draw()
    square = Square()
    square.draw()

    # obj = Drawable() # TypeError: Can't instantiate abstract class Drawable
    ```

#### 3\. Abstract Class

  * **Definition:** An abstract class is a class that cannot be instantiated on its own. It's meant to be subclassed, and it can contain both abstract methods (methods declared without an implementation) and concrete methods (methods with full implementation).

  * **Characteristics:**

      * Cannot be instantiated directly.
      * Can have one or more abstract methods, which must be implemented by any concrete subclass.
      * Can also have concrete methods and instance variables.
      * Used as a blueprint for other classes, enforcing that certain methods *must* be provided by subclasses.

  * **Purpose:** To define a common interface and potentially some common behavior for a group of related classes, while deferring the implementation of specific details to its concrete subclasses. It often represents a partial implementation of a concept.

  * **Example (Python using `abc` module):**

    ```python
    from abc import ABC, abstractmethod

    class Vehicle(ABC): # Abstract Class
        def __init__(self, make, model):
            self.make = make
            self.model = model

        @abstractmethod
        def start_engine(self): # Abstract method - must be implemented by subclasses
            pass

        def display_info(self): # Concrete method
            print(f"Make: {self.make}, Model: {self.model}")

    class Car(Vehicle): # Concrete subclass
        def __init__(self, make, model, num_doors):
            super().__init__(make, model)
            self.num_doors = num_doors

        def start_engine(self): # Implementation of the abstract method
            print(f"The {self.make} {self.model} car's engine starts with a key.")

    class Motorcycle(Vehicle): # Concrete subclass
        def __init__(self, make, model, has_sidecar):
            super().__init__(make, model)
            self.has_sidecar = has_sidecar

        def start_engine(self): # Implementation of the abstract method
            print(f"The {self.make} {self.model} motorcycle's engine roars to life.")

    # vehicle = Vehicle("Generic", "Model") # TypeError: Can't instantiate abstract class Vehicle

    my_car = Car("Toyota", "Camry", 4)
    my_car.display_info()
    my_car.start_engine()

    my_motorcycle = Motorcycle("Harley-Davidson", "Iron 883", False)
    my_motorcycle.display_info()
    my_motorcycle.start_engine()
    ```
