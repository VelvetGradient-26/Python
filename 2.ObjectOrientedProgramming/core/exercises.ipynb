```python
# Circle Class
from math import pi
from random import randint

class Circle: 
    def __init__(self, radius): 
        self.radius = radius
        self.diameter = radius * 2

    def circumference(self): 
        return (2 * pi * self.radius)
    
    def area(self): 
        return (2 * pi * self.radius * self.radius)
    
circle = Circle(randint(10, 100))
print(f"Radius: {circle.radius}")
print(f"Circumference: {circle.circumference()}")
print(f"Area: {circle.area()}")
```

```python
# Employee Class

class Employee: 
    employee_ID = 101
    employee_count = 0

    def __init__(self, name, salary, designation):
        self.name = name
        self.__salary = salary
        self.emp_id = "E" + str(Employee.employee_ID)
        self.designation = designation

        Employee.employee_ID += 1
        Employee.employee_count += 1

    def show_details(self): 
        print(f"Name: {self.name}")
        print(f"Employee ID: {self.emp_id}")
        print(f"Designation: {self.designation}")

    @classmethod
    def get_employee_count(cls): 
        return cls.employee_count
    
employee = Employee("Deepak", salary=2500000, designation="CEO")
employee2 = Employee("Prateek", salary=26000, designation="CyberSec")

print(employee.emp_id)
print(employee2.emp_id)

print(Employee.get_employee_count())

    
```
