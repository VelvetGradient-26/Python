```python
# Basic class with a constructor

class Apple:
    def __init__(self, color, flavor):
        self.color = color
        self.flavor = flavor

    # controls what is printed when the print function is used on the object
    def __str__(self):
        return f"An apple which is {self.color} in color and {self.flavor} in taste"

honeycrisp = Apple('Red', 'Sweet')
fuji = Apple("Red", "Tart")

print(honeycrisp)
print(fuji)
```

```python
class Person: 
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return (f"My name is {self.name} and My age is {self.age}")
    
    # str method overrides this, so this method can only be used without the str method or in a debugger
    def __repr__(self):
        return f"<Person: {self.name}, {self.age}>"

person = Person("Deepak", 19)

print(person.__repr__())
```

```python
# Rectangle class
class Rectangle: 
    # class variable
    count = 0
    # self is the reference to the current object
    def __init__(self, length, breadth): 
        # Instance variables
        self.length = length 
        self.breadth = breadth

        # class variable
        Rectangle.count += 1

    # instance methods
    def perimeter(self):
        return 2 *(self.length + self.breadth)
    
    def area(self): 
        return self.length * self.breadth
    
    # class method; should use the classmethod decorator
    @classmethod
    def get_count(cls): 
        return cls.count
    
    @staticmethod
    def odd_or_even(x): 
        return "Even" if x % 2 == 0 else "Odd"

rec = Rectangle(10, 5)
print(rec.perimeter())
print(rec.area())

rect = Rectangle(10, 20)
print(rect.perimeter())
print(rect.area())

print(Rectangle.get_count())

print(rect.odd_or_even(874))
```

```python
# Property Method

class Rectangle: 
    def __init__(self, l, b): 
        self.length = l
        self.breadth = b

    @property
    def length(self): 
        return self._length
    
    @length.setter
    def length(self, l): 
        if l >= 0: 
            self._length = l
        else: 
            self._length = 1

    @property
    def breadth(self): 
        return self._breadth
    
    @breadth.setter
    def breadth(self, b): 
        if b >= 0: 
            self._breadth = b
        else: 
            self._breadth = 1

rect = Rectangle(10, 5)
print(rect.length)
print(rect.breadth)

rect.length = -9
print(rect.length)

rect.breadth = -1
print(rect.breadth)
```

```python
# INHERITANCE

# creating a cuboid class using the Rectangle class

class Cuboid(Rectangle): 
    def __init__(self, l, b, h): 
        super().__init__(l,b)
        self.height = h

    def volume(self): 
        return self.length * self.breadth * self.height
    
cuboid = Cuboid(10,5,3)
print(cuboid.volume())
```

```python
# Polymorphism - Duck Typing

class Duck: 
    def talk(self): 
        print("Duck is Talking")

    def walk(self): 
        print("Duck is walking")

class Dog: 
    def talk(self): 
        print("Dog is talking")
    
    def walk(self): 
        print("Dog is walking")

class Horse: 
    def talk(self): 
        print("Horse is talking")
    
    def run(self): 
        print("Horse is running")

def person(pet): 
    pet.talk()
    if hasattr(pet, 'walk'):
        pet.walk()

dog = Dog()
person(dog)

duck = Duck()
person(duck)

horse = Horse()
person(horse)
        
```

```python
# Polymorphism - Method Overloading (Same function name, with different: Type of parameters, number of parameters)

# method overloading is enabled by default in python for type of parameters

# works for all data types
def sum(a, b): 
    return a + b


# the above function is shadowed and this is the new 'sum' function
def sum(a, b, c): 
    return a + b + c
```

```python
# Polymorphism - Method Overriding

class Parent: 
    def show(self): 
        print("Parent Show")

    def display(self): 
        print("I am a Parent")

class Child(Parent): 
    # this method will override the show method from the parent class
    def show(self): 
        print("Child Show")

    # you can get both of them via the super() method
    def display(self): 
        super().display()
        print("I am a Child")
    

c = Child()
c.show()

print(c.display)
```

```python
# Operator level Polymorphism 

from math import pi, pow

class Circle: 
    def __init__(self, diameter):
        self.diameter = diameter
        self.radius = diameter / 2

    def circumference(self):
        return 2 * pi * self.radius
    
    def area(self): 
        return pi * pow(self.radius, 2)
    
    def __add__(self, other): 
        return self.area() + other.area()
    
circle1 = Circle(347.67)
circle2 = Circle(123.78)
        
print(f"Area of circle 1: {circle1.area():.2f}")
print(f"Area of circle 2: {circle2.area():.2f}")

print(circle1 + circle2)
```

```python
# Add The volume of two cuboids

class Cuboid(Rectangle): 
    def __init__(self, length, breadth, height): 
        super().__init__(length, breadth)
        self.height = height

    def volume(self): 
        return self.length * self.height * self.breadth
    
    def __add__(self, other): 
        return self.volume() + other.volume()
    
cuboid1 = Cuboid(10, 20, 5)
print(cuboid1.volume())

cuboid2 = Cuboid(10, 5, 3)
print(cuboid2.volume())

print(cuboid1 + cuboid2)
```

```python
# Adding two vectors

class Vector: 
    def __init__(self, x, y): 
        self.x = x
        self.y = y

    def __add__(self, other): 
        x = self.x + other.x
        y = self.y + other.y
        return (x, y)

v1 = Vector(10, 2)
v2 = Vector(2, 3)

print(v1 + v2)
```

```python
from abc import ABC, abstractmethod

class Parent(ABC): 
    def meth1(self): 
        print("This is method 1 from Parent Class")

    @abstractmethod
    def meth2(self): 
        print("This is method 2 from Parent class")


class Child(Parent): 
    def meth3(self): 
        print("This is Child Method 3 from Child Class")

    def meth2(self): 
        print("This is method 2 from Child Class (Overidden Method)")

c = Child()
c.meth1()
c.meth2()
c.meth3()
```

The Four Pillars of Object-Oriented Programming
Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects", which can contain data in the form of fields (often known as attributes or properties) and code in the form of procedures (often known as methods). The primary purpose of OOP is to increase the flexibility and maintainability of programs. This paradigm is built upon four main principles, often called the pillars of OOP.

1. Encapsulation
Encapsulation is the practice of bundling the data (attributes) and the methods (functions) that operate on the data into a single unit, known as a class. It also involves restricting direct access to some of an object's components, which is a concept called data hiding.

The internal state of an object is protected from outside access and can only be changed through a defined set of public methods.

Benefits of Encapsulation
Data Hiding: It protects the internal state of an object from accidental modification by outside code.
Increased Flexibility: The internal implementation of a class can be changed without affecting the code that uses it.
Modularity: Each object is a self-contained unit, which makes the program easier to understand and maintain.

2. Abstraction
Abstraction means hiding the complex implementation details and showing only the necessary features of an object. It's about simplifying reality by modeling classes appropriate to the problem. The user of the object only needs to know what it does, not how it does it.

Benefits of Abstraction
Simplicity: Makes objects easier to use and reduces the complexity of the system.
Reduces Impact of Change: The internal implementation can be changed without affecting the classes that use it.
Focus: Allows developers to focus on the high-level functionality of objects rather than getting bogged down in the details.

3. Inheritance
Inheritance is a mechanism that allows a new class (the subclass or derived class) to adopt the properties and methods of an existing class (the superclass or base class). This creates an "is-a" relationship. For example, a Car is a Vehicle.

Benefits of Inheritance
Code Reusability: It promotes the reuse of code from the base class, avoiding duplication.
Extensibility: You can add new features to a subclass without modifying the base class.
Logical Hierarchy: It helps in creating a clear and logical class structure.

4. Polymorphism
Polymorphism, which means "many forms," is the ability of an object to take on many forms. In OOP, it most commonly means that a single method name can work for different types of objects. It allows you to define one interface and have multiple implementations.

Benefits of Polymorphism
Flexibility and Decoupling: Allows for writing generic code that can work with objects of different types, without needing to check the object's type.
Extensibility: New classes that implement the common interface can be added easily without changing the code that uses them.
