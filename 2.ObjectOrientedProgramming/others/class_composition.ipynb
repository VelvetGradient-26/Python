```python
# Class Composition in many ways is considered a best practice and simpler than Inheritance

class BookShelf: 
    def __init__(self, *books): 
        self.books = books

    def __str__(self): 
        return f"Bookshelf: {len(self.books)} books."
    
    def get_books(self): 
        for book in self.books: 
            print(book)

class Book: 
    def __init__(self, name): 
        self.name = name
    
    def __str__(self): 
        return f"Book: {self.name}"
    
book1 = Book("Harry Potter")
book2 = Book("Python 101")
shelf = BookShelf(book1, book2)

shelf.get_books()

print(shelf)
```

```python
# Class composition with Type hinting
from typing import List

class Book: 
    def __init__(self, name: str, page_count: int): 
        self.name = name 
        self.page_count = page_count

class Bookshelf: 
    def __init__(self, *books: List[Book]):
        self.books = books

    def __str__(self) -> str:
        return f"Bookshelf with {len(self.books)}"

```
